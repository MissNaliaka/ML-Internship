## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

In lane 2 , the goal is to rank pages by priority for review. So the natural prediction is: "Will this page decline (or continue declining) in the near future?" That gives a binary label per page: yes or no.

The label can come from a defined rule from the current data “trend_direction = down” or it can be sourced as an observed outcome measured from actual future data. For example, a page had X impressions over the next 30 days, which dropped vs. the previous 90 days.

The starter dataset uses a defined rule — trend_direction == "down" is a bucket calculated from the current window, not something that happened later. That's a proxy label: a stand-in, not the real thing.
A stronger label would be an observed outcome: features from days 1-90  ->  did this page decline over days 91-120? That's a real future event you can verify in the data.

Therefore, if using the starter dataset, the answer is: "I am predicting whether a page is declining, using the proxy label trend_direction == 'down'. This is a defined rule from current data, not a future outcome — a limitation I'll address in later weeks."

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

The baseline uses fixed thresholds like "stale if >= 180 days old" and "low CTR if < 0.5" — but these are arbitrary cutoffs. Real pages have many interacting signals: a page at position 3 with declining CTR is different from one at position 15 with the same CTR. ML can learn the optimal weight and combination of signals from data, rather than relying on hand-tuned if-statements.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Show the baseline's arbitrary thresholds
baseline_rules = {
    "stale_visible_page": "days_since_last_update >= 180 AND impressions_90d >= 500",
    "declining_with_demand": "trend_direction == 'down' AND impressions_90d >= 100",
    "thin_visible_page": "word_count < 1200 AND impressions_90d >= 250",
    "low_ctr_visible_page": "avg_position <= 20 AND ctr < 0.5"
}

for rule, definition in baseline_rules.items():
    print(f"{rule}: {definition}")
```
